# read data

Below we're reading the dataframe and displaying it using jupyter formatting, truncating the output.

```python
import os
import xarray as xr

file_path = os.path.join(os.path.dirname(os.path.realpath("__file__")),
                         '../data/Europe_2019/Austria/MERRA2_400.tavg1_2d_slv_Nx.20190101.nc4')

with xr.open_mfdataset(file_path, combine='by_coords') as ds_wind:
    df_wind = ds_wind.to_dataframe()

df_wind
```

Understand what columns mean from here: https://gmao.gsfc.nasa.gov/pubs/docs/Bosilovich785.pdf (page 23).

Next we are showing the full dataframe output, without truncating it.

```python
import pandas as pd

with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(df_wind)
```

You can see that the data is ~2300 rows in size and contains information for each hour of the day, starting at 00:30, and going to 23:30.
